```python
import numpy as np
import pandas as pd
```

```python
maturities = [1, 2, 3, 5, 7, 10, 20, 30]
yields = []

for mat in maturities:
    t_yield = pd.read_csv('DGS' + str(mat) + '.csv', index_col=0)
    t_yield.index = pd.to_datetime(t_yield.index)
    yields.append(t_yield)
```

```python
y_curve = pd.concat(yields, axis=1)
```

```python
y_curve[y_curve.isna().any(axis=1)]
```

```python
# No NaN from year 2010 to 2020
y_curve = y_curve[(y_curve.index.year >= 2010) & (y_curve.index.year < 2021)]
```

```python
y_curve
```

```python
# Remove '.' in dataframe
y_curve = y_curve[~y_curve.isin(['.']).any(axis=1)]
```

```python
y_curve
```

```python
y_curve = y_curve.astype(float)
chg_yield = y_curve.diff()
```

```python
chg_yield
```

```python
chg_yield.corr()
```

```python
eig_vals, eig_vects = np.linalg.eig(chg_yield.corr())
```

```python
eig_vects = pd.DataFrame(eig_vects)
eig_vals = pd.DataFrame(eig_vals, columns=['eigenval'])
```

```python
eig_vects
```

```python
eig_vals['explained_prop'] = eig_vals['eigenval'] / eig_vals['eigenval'].sum()
eig_vals.style.format({'explained_prop': '{:.2%}'})
```
